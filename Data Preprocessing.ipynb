Data Preprocessing

```python
#import JPY/USD data. JPY-raw data in giving in the file and is download from WIND.
import pandas as pd
import numpy as np
JPY_whole=pd.read_excel('JPY-raw.xlsx',index_col=0)
JPY_whole.head(5)
```

```python
#We only use recent years data for forecasting. The period is from 2016-01-04 to 2018-04-10, with total data number 592.
JPY=JPY_whole.loc['2016':'2018']
JPY.shape
```

```python
# print the graph of the exchange rate
import matplotlib.pyplot as plt
import seaborn as sns
JPY.plot(figsize=(12,8))
plt.legend(bbox_to_anchor=(1.25, 0.5))
plt.title("JPY/USD exchange rate")
sns.despine()
plt.show()
```

```python
# For the time series data may be unstable, we calculate the first order and second order difference. 
# We plot all of them in the graph. 
# There is no significant improvement in first order difference and second order difference, we therefore use diff_1 .
JPY['diff_1'] = JPY['JPY/USD'].diff(1)
JPY['diff_2'] = JPY['diff_1'].diff(1)
JPY.plot(subplots=True, figsize=(18, 12))
```
